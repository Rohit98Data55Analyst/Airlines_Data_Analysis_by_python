# Q.2. Show Bar graph representing the departure time and arrival time 

```python
file.head()
```

```python
departure = file['departure_time'].value_counts()
```

```python
arrival = file['arrival_time'].value_counts()
```

```python
# Showing the departure time and arrival time with their counts
fig, axs = plt.subplots(1,2,figsize = (15,4))


bar1 = axs[0].barh(departure.index , departure.values ,color=['orange','skyblue'])

axs[0].set_title("Departure time With Frequency")

for bar in bar1:
    width = bar.get_width()
    axs[0].text(width, bar.get_y() + bar.get_height()/2, str(width),
                ha='left', va='center')  
    
bar2 = axs[1].barh(arrival.index , arrival.values,color=['orange','skyblue'])

axs[1].set_title("Arrival time With Frequency")

for bar in bar2:
    width = bar.get_width()
    axs[1].text(width, bar.get_y() + bar.get_height()/2, str(width),
                ha='left', va='center')  
plt.show()
```

# Most of the flights are departure and arrival in the morning .

# Q.3. Show Bar Graphs representing the Source City and Destination City.

```python
file.head()
```

```python
source_city = file['source_city'].value_counts()
destination_city = file['destination_city'].value_counts()
```

```python
fig , ax = plt.subplots(1,2,figsize = (10,4))

bar1 = ax[0].bar(source_city.index ,source_city.values,color =['yellow','brown'])
ax[0].tick_params(axis='x', rotation=45) # to rotate the labels in x axis
for bar in bar1: # To add bar labels
    height = bar.get_height()
    ax[0].text(
               bar.get_x() + bar.get_width()/2,
               height,
               str(height),
               ha ='center' , va = 'bottom'
    )
ax[0].set_title('Departure City')    

bar2 = ax[1].bar(destination_city.index ,destination_city.values,color =['yellow','brown'])
ax[1].tick_params(axis='x', rotation=45)
for bar in bar2: # To add bar labels
    height = bar.get_height()
    ax[1].text(
               bar.get_x() + bar.get_width()/2,
               height,
               str(height),
               ha ='center' , va = 'bottom'
    )
ax[1].set_title('Arrival City')    
plt.show()
```

# Insight: According to these two bar graph most of the flights are departure from 'Delhi' and arrival in 'Mumbai'.

# Q.4. Does price varies With airlines ?

```python
airline_price = file.groupby(['airline'],as_index = False)['price'].mean().sort_values(by='price',ascending = False)
airline_price
```

```python
ax = sns.barplot(data = airline_price , x = 'airline' , y = 'price' ,palette = 'Set2' )

for bars in ax.containers:
    ax.bar_label(bars , fmt = '%1.0f')
plt.xticks(rotation = 45)    
plt.title('Price varying')
plt.show()
```

```python
# Drawing a Categorical plot showing the Mean Ticket price for each Airline
ax = sns.catplot(x ='airline' , y ='price' , kind = 'bar' , palette ='rocket' , data = file ,hue='class')

plt.show()
```

# According to above graph only 'Vistara' and 'Air_India' provide business class and their mean price also high.
